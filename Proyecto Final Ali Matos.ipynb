Graficar las MA (Medias Moviles) MA50 y MA200 para GM

```python
from matplotlib.pyplot import figure
import matplotlib.dates as mdates
from matplotlib.ticker import (MultipleLocator,AutoMinorLocator)

def moving_average(a, n):
    return np.convolve(a, np.ones(n), 'valid') / n

date = mdates.datestr2num(cdata3.Date.to_numpy())
gm_open = data3.Open.to_numpy()
inic = np.empty(49)
inic [:] = np.nan
mv50 = np.append (inic,np.array (moving_average(gm_open,50)))
inic = np.empty(99)
inic [:] = np.nan
mv100 = np.append (inic,np.array (moving_average(gm_open,100)))

fmt_month = mdates.MonthLocator(interval=1)
fig , ax = plt.subplots()
ax.plot (date, gm_open, color='b',label = 'OPEN')
ax.plot (date, mv50, color='orange', label = 'MV50')
ax.plot (date, mv100, color ='r', label = 'MV100')
ax.xaxis.set_major_locator(fmt_month)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xlabel ('Mes')
plt.title ('Medias Móviles')
plt.legend()
fig.autofmt_xdate()
plt.show()
```

### Densidad de Probabilidad

Función:

![image](Gaussiana.png)

```python
# Definir medias

mu_tsla = data.Retorno.mean()
mu_ford = data2.Retorno.mean()
mu_gm = data3.Retorno.mean()
print(f"{mu_tsla} {mu_ford} {mu_gm}")
```

```python
# Definir Desviación estandar
sigma_tsla = data.Retorno.std()
sigma_ford = data2.Retorno.std()
sigma_gm = data3.Retorno.std()
print(f"{sigma_tsla} {sigma_ford} {sigma_gm}")
```

```python
# Funciones de densidad
num_bins = 50
tsla_hist = data.Retorno.dropna().to_numpy()
ford_hist = data2.Retorno.dropna().to_numpy()
gm_hist = data3.Retorno.dropna().to_numpy()
fig, ax = plt.subplots()
n_tsla, bins_tsla = np.histogram(tsla_hist, num_bins, density=True)
n_ford, bins_ford = np.histogram(ford_hist, num_bins, density=True)
n_gm, bins_gm = np.histogram(gm_hist, num_bins, density=True)

y_tsla = ((1 / (np.sqrt(2 * np.pi) * sigma_tsla)) * np.exp(-0.5 * (1 / sigma_tsla * (bins_tsla - mu_tsla))**2))
y_ford = ((1 / (np.sqrt(2 * np.pi) * sigma_ford)) * np.exp(-0.5 * (1 / sigma_ford * (bins_ford - mu_ford))**2))
y_gm = ((1 / (np.sqrt(2 * np.pi) * sigma_gm)) * np.exp(-0.5 * (1 / sigma_gm * (bins_gm - mu_gm))**2))
ax.plot(bins_tsla, y_tsla, color='g', label = 'Tesla')
ax.plot(bins_ford, y_ford, color='r', label = 'Ford')
ax.plot(bins_gm, y_gm, color='b', label = 'GM')
plt.legend()
ax.set_xlabel('Retorno')
ax.set_ylabel('Densidad de Probabilidad')

fig.tight_layout()
plt.show()
```
